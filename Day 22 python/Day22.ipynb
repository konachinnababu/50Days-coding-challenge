```python
order_id = "ORD12345"      # string (text)
customer_name = "Rahul"    # string
product_price = 799.99     # float (decimal number)
quantity = 3               # integer
discount = 0.10            # float
is_member = True           # boolean (True/False)
```

```python
print(type(order_id))
print(type(customer_name))
print(type(product_price))
print(type(quantity))
print(type(discount))
print(type(is_member))
```

```python
Total_price = product_price * quantity
print("Total price before discount:", total_price)
```

```python
Discount_amount = total_price * discount
final_price = total_price - discount_amount

print("Discount amount:", discount_amount)
print("Total price after discount:", final_price)
```

```python
Gst = 0.12

gst_amount = final_price * gst
final_bill = final_price + gst_amount

print("GST amount:", round(gst_amount, 0))
print("Final bill amount:", round(final_bill, 0))
```

```python
print("Order ID:", order_id)
print("Customer:", customer_name)
print("Total Price:", round(total_price, 2))
print("Discount Applied:", "Yes" if discount > 0 else "No")
print("Final Amount (including GST):", round(final_bill, 2))
```
