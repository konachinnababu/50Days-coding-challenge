```python
rides = (
    (101, "Ravi", "Bangalore", 250, 4.5),
    (102, "Anjali", "Hyderabad", 180, 4.2),
    (103, "Kiran", "Bangalore", 300, 4.8),
    (104, "Meena", "Chennai", 220, 4.0),
    (105, "Arjun", "Bangalore", 150, 3.9)
)
```

```python
print(len(rides))
```

```python
for ride in rides:
    if ride[2] == "Bangalore":
        print(ride)
```

```python
total = 0

for ride in rides:
    if ride[2] == "Bangalore":
        total += ride[3]

print(total)
```

```python
highest = max(rides, key=lambda x: x[4])
print(highest)
```

```python
city_data = {}

for ride in rides:
    city = ride[2]
    rating = ride[4]

    if city not in city_data:
        city_data[city] = [rating]
    else:
        city_data[city].append(rating)

# average calculate
for city in city_data:
    avg = sum(city_data[city]) / len(city_data[city])
    city_data[city] = round(avg, 1)

print(city_data)
```
